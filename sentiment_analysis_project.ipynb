```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob
```

```python
data = {
    "Review": [
        "This product is amazing and I love it",
        "The product is excellent and very useful",
        "I am very happy with this purchase",
        "The quality is very good",
        "This is the worst product I have ever bought",
        "I hate this product",
        "Very poor quality and bad service",
        "The product stopped working after one day",
        "The product is okay",
        "It is an average product",
        "Nothing special about this product",
        "The product is available"
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
print("Number of reviews:", len(df))
print("\nDataset Information:")
print(df.info())
```

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["Review"].apply(get_sentiment)

df
```

```python
sentiment_counts = df["Sentiment"].value_counts()

print(sentiment_counts)
```

```python
plt.figure(figsize=(8,5))

df["Sentiment"].value_counts().plot(kind="bar")

plt.title("Sentiment Analysis of Customer Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["Sentiment"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Percentage of Sentiments")
plt.ylabel("")

plt.show()
```

```python
def detect_emotion(text):
    text = text.lower()

    if any(word in text for word in ["amazing", "excellent", "happy", "love"]):
        return "Joy"
    elif any(word in text for word in ["worst", "hate", "poor", "bad"]):
        return "Anger/Sadness"
    elif any(word in text for word in ["okay", "average", "nothing special"]):
        return "Neutral"
    else:
        return "Other"
```

```python
df["Emotion"] = df["Review"].apply(detect_emotion)

df
```

```python
plt.figure(figsize=(8,5))

df["Emotion"].value_counts().plot(kind="bar")

plt.title("Emotion Analysis of Reviews")
plt.xlabel("Emotion")
plt.ylabel("Number of Reviews")

plt.show()
```
